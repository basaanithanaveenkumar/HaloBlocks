# 👼 HaloBlocks Tutorial

Welcome to the official tutorial for **HaloBlocks**! 

HaloBlocks is a modern, high-performance Python library for modular neural network components. In this notebook, we'll explore how to build and compose models using our three primary interfaces:
1. **Layers API**: Direct class instantiation.
2. **Factory API**: Configuration-driven creation.
3. **Composite Blocks**: Building sequential pipelines.

## ⚙️ Setup

First, let's ensure we have everything installed and our environment set up.

```python
# Install if needed
# !pip install haloblocks torch

import torch
import haloblocks as hb
import haloblocks.layers as layers

print(f"HaloBlocks version: {hb.__version__ if hasattr(hb, '__version__') else '0.1.0'}")
```

## 🧩 Part 1: Using `haloblocks.layers` (The Keras Style)

This is the most direct way to use HaloBlocks. You can access any registered block as if it were a class in the `layers` module.

```python
# Create a Multi-Head Attention layer
mha = layers.multi_head_attn(emb_dim=256, num_heads=8, causal_mask=True)

print(f"Created layer: {mha.name}")
print(f"Configuration: {mha}")
```

```python
# Run a dummy forward pass
x = torch.randn(2, 64, 256)
output = mha(x)
print(f"Input shape: {x.shape}")
print(f"Output shape: {output.shape}")
```

## 🧱 Part 2: The Configurable MLP

The `MLP` block is highly versatile, supporting custom activations, bias control, and flexible output behavior.

```python
# Create an MLP with GELU and no bias on the last layer
mlp = layers.mlp(
    input_dim=128, 
    hidden_dims=[256, 256], 
    output_dim=64,
    activation='gelu',
    bias=True,
    last_layer_activation=False  # Final layer is a linear projection
)

x = torch.randn(4, 128)
out = mlp(x)
print(f"MLP output shape: {out.shape}")
print(f"MLP structure: {mlp}")
```

## 🛠️ Part 3: Using the `create` Function

The `hb.create()` function is a powerful tool that allows you to instantiate blocks using string names and keyword arguments. This is ideal for quickly swapping components.

```python
# Create a Scaled Dot Product Attention block
attn = hb.create('scaled_dot_product_attention', dropout=0.1)

# For SDPA, we need Q, K, V
q = torch.randn(2, 8, 64, 32)
k = torch.randn(2, 8, 64, 32)
v = torch.randn(2, 8, 64, 32)

out = attn(q, k, v)
print(f"SDPA output shape: {out.shape}")
```

## 🏗️ Part 4: Configuration-Driven Design

The `create` function also accepts dictionaries, making it perfect for loading models from JSON or YAML files.

```python
config = {
    'type': 'self_attention_basic',
    'emb_dim': 128,
    'return_attn_weights': True
}

block = hb.create(config)
x = torch.randn(1, 10, 128)
out, weights = block(x)

print(f"Attention Weights shape: {weights.shape}")
```

## 🧱 Part 5: Composite Blocks

HaloBlocks makes it easy to stack blocks together. You can even define these stacks in your configuration!

```python
composite_config = {
    'type': 'composite_block',
    'blocks': [
        {'type': 'multi_head_attn', 'emb_dim': 128, 'num_heads': 4},
        {'type': 'mlp', 'input_dim': 128, 'hidden_dims': [256, 128]}
    ]
}

pipeline = hb.create(composite_config)
print(pipeline)
```
